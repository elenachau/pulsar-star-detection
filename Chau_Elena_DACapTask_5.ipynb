```python
col_names = [
    'Mean of the integrated profile',
    'Standard deviation of the integrated profile',
    'Excess kurtosis of the integrated profile',
    'Skewness of the integrated profile',
    'Mean of the DM-SNR curve',
    'Standard deviation of the DM-SNR curve',
    'Excess kurtosis of the DM-SNR curve',
    'Skewness of the DM-SNR curve',
    'Class'
]

df.columns = col_names
```

```python
# class_col = df.iloc[1:,-1]
dm_snr_curve_skew = df.iloc[1:,-2]
excess_kurtosis_DM_SNR_curve = df.iloc[1:,-3]

# print(f'Class: {sps.skew(class_col)}')
print(f'DM-SNR curve skewness: {sps.skew(dm_snr_curve_skew)}')
print(f'Excess kurtosis of the DM-SNR curve: {sps.skew(excess_kurtosis_DM_SNR_curve)}')
print(f'Mean of the integrated profile: {sps.skew(df["Mean of the integrated profile"])}\n')
print(f'Standard deviation of the integrated profile: {sps.skew(df["Standard deviation of the integrated profile"])}')
print(f'Excess kurtosis of the integrated profile: {sps.skew(df["Excess kurtosis of the integrated profile"])}')
print(f'Skewness of the integrated profile: {sps.skew(df["Skewness of the integrated profile"])}')
print(f'Mean of the DM-SNR curve: {sps.skew(df["Mean of the DM-SNR curve"])}')
print(f'Standard deviation of the DM-SNR curve: {sps.skew(df["Standard deviation of the DM-SNR curve"])}')
```
